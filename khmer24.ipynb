```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs
```

```python
def get_webpage(url):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(url,headers=hdr)
    page = urlopen(req)
    return bs(page,"html.parser")
    
```

```python
test= get_webpage('https://www.khmer24.com/')
```
